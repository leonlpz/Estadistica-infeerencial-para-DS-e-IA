```python
import pandas as pd
import numpy as np 
import seaborn as sns 
import scipy.stats as st
```

```python
np.random.seed(40) # distribucion creada
var1 = np.random.randint(0, 20, 20) # Días de vacaciones
var2 = var1+np.random.normal(0, 1, 20) # Dinero gastado
```

```python
var1
```

```python
var2
```

```python
grafico = sns.regplot(x=var1, y=var2, ci=85) # vizualizacion de resultados con intervalo de 85%
```

```python
# Unimos las variables var1 y var2 (estamos creando un tuple) a partir de zip
tuple = list(zip(var1, var2))
tuple
```

```python
# Transformamos el tuple a un data frame a partir de DataFrame
tabla = pd.DataFrame(tuple,
                  columns = ['Días_vacaciones', 'Dinero_gastado'])
tabla
```

```python
# Calcula los intervalos de confianza a un 95% para ambas variables
st.t.interval(confidence=0.95, df=len(tabla)-1, loc=np.mean(tabla), scale=st.sem(tabla))
```
