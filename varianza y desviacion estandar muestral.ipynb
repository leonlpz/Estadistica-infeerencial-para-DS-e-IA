```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import csv
```

### Lectura de la base de datos 
https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data

```python
#url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.csv"
names = ['sepal-length', 'sepal-width', 'petal-length', 'petal-width', 'class']
iris = pd.read_csv('databasesirisdata.txt', sep=',', names=names)
iris.head() # Exploración de la tabla , Largo y ancho del sépalo y de los pétalos
```

### Información extra acerca de petalos y sepalos:
 https://www.sciencefacts.net/sepals.html

```python
# Estudio de la distribución de la variable 
plt.hist(iris['sepal-length'], color = 'green', bins = int(180/5))
plt.xlabel('Sepal Length')
```

```python
sns.histplot(iris['sepal-length'], kde=True, bins=int(180/5), color='green', alpha=0.5)
plt.show # el grafico de kernel nos mostrara si efectivamente es una distribución normal
```

## Calculo de varianza

$$\sigma^{2} = \frac{\sum_{i=1}^{n} 
  \left(x_{i} - \mu\right)^{2}}
  {N}$$

```python
iris["sepal-length"].var()
```

## Calculo de desviación estandar

$$\sigma = \sqrt{\sigma}^{2} = \sqrt{\frac{\sum\limits_{i=1}^{n} \left(x_{i} - \mu\right)^{2}} {N}}$$

```python
iris["sepal-length"].std()
```

## Calculo de la media

```python
iris["sepal-length"].mean()
```

## Interpretación de datos poblacionales:
### Con el indicador de la media y su respectivas desviación estandar y varianza podemos notar que son un valor significativo (elevadas) correspondiente al 14% del promedio. Por lo tanto tenemos valores de longitud de sepalos poco unidos y muy separados. 

| Coeficiente de Variación | Apreciación de la Muestra |
| --- | --- |
| 0% a 10% | Muy homogénea |
| 11% a 15% | Homogénea |
| 16% a 25% | Heterogénea |
| 26% o más | Muy heterogénea |

## Ahora se crea una muestra mediante muestreo aleatorio simple

```python
muestra = iris.sample(frac=0.5) # 50% de los datos
muestra
```

## Calculo de la varianza, desviación estandar y media de la muestra

```python
muestra["sepal-length"].var()
```

```python
muestra["sepal-length"].std()
```

```python
muestra["sepal-length"].mean()
```

### Asi como con los datos de la población general, el indicador de la media y su respectivas desviación estandar y varianza de la muestra, podemos notar que son un valor significativo (elevadas) respecto del promedio.
